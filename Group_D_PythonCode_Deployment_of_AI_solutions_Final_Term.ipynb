```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML, FileLink

def data_analysis(data):
    # Display basic statistics
    display(HTML("<h3 style='color: #3366cc;'>Basic Statistics:</h3>"))
    display(data.describe())

    # Display a pair plot
    display(HTML("<h3 style='color: #3366cc;'>Pair Plot:</h3>"))
    pd.plotting.scatter_matrix(data, figsize=(12, 12), marker='o', hist_kwds={'bins': 10}, alpha=0.5)
    plt.show()

    # Correlation Heatmap
    display(HTML("<h3 style='color: #3366cc;'>Correlation Heatmap:</h3>"))
    correlation_matrix = data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
    plt.show()

    # Boxplot for each feature
    display(HTML("<h3 style='color: #3366cc;'>Boxplots:</h3>"))
    data.boxplot(figsize=(10, 8), rot=45)
    plt.show()

    # Distribution of Age
    display(HTML("<h3 style='color: #3366cc;'>Distribution of Age:</h3>"))
    sns.histplot(data['Age'], bins=20, kde=True)
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()

# Function to handle file upload
def handle_file_upload():
    uploaded_file = None

    try:
        from google.colab import files
        display(HTML("<h3 style='color: #3366cc;'>Upload a CSV file for analysis:</h3>"))
        uploaded = files.upload()
        uploaded_file = list(uploaded.keys())[0]

        # Check if the uploaded file is a CSV file
        if not uploaded_file.endswith('.csv'):
            raise ValueError("Please upload a CSV file.")

        # Display success message in green
        display(HTML("<h2 style='color: #4CAF50;'>File uploaded successfully!</h2>"))

    except ImportError:
        print("Please run this code in a Google Colab environment.")

    return uploaded_file

# Heading for the project
display(HTML("<h1 style='color: #3366cc;'>Data Analysis Project</h1>"))

# Allow users to upload a CSV file
file_path = handle_file_upload()

# Perform data analysis if data is provided
if file_path:
    data = pd.read_csv(file_path)
    data_analysis(data)
else:
    display(HTML("<h2 style='color: #3366cc;'>No data provided for analysis.</h2>"))
```
